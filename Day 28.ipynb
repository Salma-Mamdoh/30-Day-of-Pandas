##  Students and Examinations Leetcode Task28

```python
import pandas as pd
import numpy as np
```

```python
data = [[1, 'Alice'], [2, 'Bob'], [13, 'John'], [6, 'Alex']]
Students = pd.DataFrame(data, columns=['student_id', 'student_name']).astype({'student_id':'Int64', 'student_name':'object'})
data = [['Math'], ['Physics'], ['Programming']]
Subjects = pd.DataFrame(data, columns=['subject_name']).astype({'subject_name':'object'})
data = [[1, 'Math'], [1, 'Physics'], [1, 'Programming'], [2, 'Programming'], [1, 'Physics'], [1, 'Math'], [13, 'Math'], [13, 'Programming'], [13, 'Physics'], [2, 'Math'], [1, 'Math']]
Examinations = pd.DataFrame(data, columns=['student_id', 'subject_name']).astype({'student_id':'Int64', 'subject_name':'object'})
```

```python
Examinations
```

```python
df2=Examinations.groupby(['student_id','subject_name']).size().reset_index(name='attended_exams')
df2
```

```python
all_combinations = pd.MultiIndex.from_product([Students['student_id'], Subjects['subject_name']], names=['student_id', 'subject_name'])
all_combinations_df = pd.DataFrame(index=all_combinations).reset_index()
all_combinations_df
```

```python
# Left join the grouped data with all possible combinations
result = all_combinations_df.merge(df2, on=['student_id', 'subject_name'], how='left').fillna(0)
result
```

```python
# Add student names to the result
result = result.merge(Students, on='student_id', how='left')
result
```

```python
# Order the result table by student_id and subject_name
result = result.sort_values(by=['student_id', 'subject_name'])
result[['student_id', 'student_name', 'subject_name', 'attended_exams']]
```
